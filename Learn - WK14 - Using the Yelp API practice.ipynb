```python
#Import Classes
import json
from yelpapi import YelpAPI
import pandas as pd
```

```python
with open('/Users/jonme/.secret/yelp_api.json') as f: #change the path to match YOUR path!!
    login = json.load(f)
login.keys()
```

```python
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
# use our yelp_api variable's search_query method to perform our API call
search_results = yelp_api.search_query(location='Minneapolis, MN',
                                       term='brewery')
print(type(search_results))
print(search_results.keys())
search_results['total']
```

```python
#Search results to a Dataframe
biz = pd.DataFrame(search_results['businesses'])
biz.head(2) 
```

```python
len(search_results['businesses'])
```

```python
search_results = yelp_api.search_query(location='Minneapolis, MN',
                                       term='brewery', offset=20)
biz20 = pd.DataFrame(search_results['businesses'])
biz20.head(2)
```

```python
## concatenate the previous results and new results. 
businesses = pd.concat([biz, biz20],
                      ignore_index=True)
display(businesses.head(3), businesses.tail(3))
```

```python
#convert to JSON
biz_json = businesses.to_json(orient='records')
```

```python
# Save it to a file
with open('Learn - WK14 - Yelp practive.json','w') as file:
    json.dump(biz_json,f)
```
